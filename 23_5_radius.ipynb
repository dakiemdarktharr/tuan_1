```python
import folium
trung_tam = [10.760500, 106.661500]
m = folium.Map(location=trung_tam, zoom_start=12)
folium.Marker(location=trung_tam, popup="Trung tam phan phoi", icon=folium.Icon(color='red', icon='truck', prefix='fa')).add_to(m)
folium.Circle(location=trung_tam, radius=3000, color='green', fill=True, fill_opacity=0.2, popup='Vung 3km').add_to(m)
folium.Circle(location=trung_tam, radius=5000, color='orange', fill=True, fill_opacity=0.15, popup='Vung 5km').add_to(m)
folium.Circle(location=trung_tam, radius=10000, color='red', fill=True, fill_opacity=0.1, popup='Vung 10km').add_to(m)
display(m)
```
